## Interpolation methods
https://www.youtube.com/watch?v=w0h8eEe97dk&t=883s

```python
import numpy as np
from scipy import interpolate
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (12,6)
```

```python
a_pts = np.linspace(0, 2*np.pi, 10)
b_pts = np.sin(a_pts)
a_vals = np.linspace(0, 2*np.pi, 50)
b_vals = np.interp(a_vals, a_pts, b_pts)
```

```python
print(len(a_pts),'\n')
print(len(a_vals),'\n')
print(len(b_pts),'\n')
print(len(b_vals),'\n')
```

```python
plt.plot(a_pts, b_pts, 'o')
plt.plot(a_vals, b_vals, 'k-.')
plt.title('Linear Interpolation')
plt.grid(True, which='both')
plt.xticks(np.arange(0, 7, step=0.5))
plt.show()
```

```python
x_pts = np.linspace(0, 2*np.pi, 10)
y_pts = np.sin(x_pts)
splines = interpolate.splrep(x_pts, y_pts)
x_vals = np.linspace(0, 2*np.pi, 50)
z_vals = interpolate.splev(x_vals, splines)
```

```python
plt.plot(x_pts, y_pts, 'o')
plt.plot(x_vals, y_vals, 'k-.')
plt.title('Cubic Spline Interpolation')
plt.grid(True, which='both')
plt.xticks(np.arange(0, 7, step=0.5))
plt.show()
```

```python
x_pts = np.linspace(0, 2*np.pi, 10)
d_pts = np.sin(a_pts)
x_vals = np.linspace(0, 2*np.pi, 50)
d_vals = np.interp(a_vals, x_pts, d_pts)
```

```python
plt.plot(b_vals, 'b-')
plt.plot(y_vals, 'k-.')
plt.title('Linear vs. Cubic Spline Interpolation')
plt.grid(True, which='both')
plt.show()
```
